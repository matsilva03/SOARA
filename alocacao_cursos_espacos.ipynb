```python
class TimeSlot:
    """Handles time slot management for course scheduling"""
    
    @staticmethod
    def create_time_slots(day: str, time_range: str) -> List[int]:
        # Existing time slot creation code remains the same
        time_slots = [0] * 48
        
        day_to_base_index = {
            "Segunda": 0,
            "Terça": 8,
            "Quarta": 16,
            "Quinta": 24,
            "Sexta": 32,
            "Sábado": 40
        }
        
        time_to_index = {
            "17h20": 0,
            "18h10": 1,
            "19h00": 2,
            "19h50": 3,
            "20h50": 4,
            "21h40": 5,
            "22h30": 6
        }
        
        start_time, end_time = time_range.split('-')
        base_idx = day_to_base_index.get(day, 0)
        start_idx = base_idx + time_to_index.get(start_time, 0)
        
        time_range_slots = {
            "19h00": 2,
            "20h40": 2,
            "22h30": 2,
        }
        
        if time_range == "19h00-22h30":
            slots_needed = 4
        else:
            slots_needed = time_range_slots.get(end_time, 2)
        
        for i in range(start_idx, start_idx + slots_needed):
            if i < len(time_slots):
                time_slots[i] = 1
                
        return time_slots

    @staticmethod
    def has_conflict(slots1: List[int], slots2: List[int]) -> bool:
        return any(s1 and s2 for s1, s2 in zip(slots1, slots2))
```

# 3. Classe de Alocação de Salas

```python
class ClassroomAssignment:
    def __init__(self, courses_data: List[Dict], rooms_data: List[Dict]):
        self.courses_data = courses_data
        self.rooms_data = rooms_data.copy()
        self.weights = {
            'floor_pref': 10,
            'lab_usage': 5,        # Penalidade para usar lab quando não necessário
            'wrong_room': 20,      # Penalidade para usar sala quando precisa de lab
            'distance': 2,
            'capacity_penalty': 15
        }
        self._calculate_floor_matches()
            
    def _calculate_floor_matches(self):
        """Calculate floor_match dynamically for each room based on course preferences"""
        for room in self.rooms_data:
            room['floor_match_dict'] = {}
            for course in self.courses_data:
                pref_floor = course['pref_floor']
                
                if pref_floor <= 0:
                    room['floor_match_dict'][course['name']] = 1
                else:
                    room['floor_match_dict'][course['name']] = 1 if room['floor'] == pref_floor else 0
        
    def create_model(self) -> Tuple[LpProblem, Dict]:
        """Creates and returns the optimization model with flexible lab allocation"""
        model = LpProblem("Classroom_Assignment", LpMinimize)
        
        # Sets
        D = range(len(self.courses_data))
        S = [r for r, data in enumerate(self.rooms_data) if data['type'] == 'sala']
        L = [r for r, data in enumerate(self.rooms_data) if data['type'] == 'lab']
        R = range(len(self.rooms_data))
        H = range(48)
        
        # Decision Variables
        x = LpVariable.dicts("assign",
                           ((d, r) for d in D for r in R),
                           cat='Binary')
        
        y = LpVariable.dicts("distance",
                           (d for d in D),
                           lowBound=0,
                           cat='Integer')
        
        # Capacity violation variables
        z = LpVariable.dicts("capacity_violation",
                           ((d, r) for d in D for r in R),
                           lowBound=0,
                           cat='Integer')
        
        # Variables for tracking room type mismatches
        w_lab = LpVariable.dicts("wrong_lab_usage",
                              (d for d in D),
                              cat='Binary')  # 1 if course needs lab but uses regular room
        
        w_sala = LpVariable.dicts("wrong_sala_usage",
                               (d for d in D),
                               cat='Binary')  # 1 if course needs regular room but uses lab
        
        # Objective Function
        model += (
            self.weights['floor_pref'] * lpSum(x[d,r] * self.courses_data[d]['course_floor_pref'] * 
                                             (1 - self.rooms_data[r]['floor_match_dict'][self.courses_data[d]['name']])
                                             for d in D for r in R) +
            self.weights['lab_usage'] * lpSum(w_sala[d] for d in D) +  # Penalidade por usar lab quando não necessário
            self.weights['wrong_room'] * lpSum(w_lab[d] for d in D) +  # Penalidade maior por usar sala quando precisa de lab
            self.weights['distance'] * lpSum(y[d] for d in D) +
            self.weights['capacity_penalty'] * lpSum(z[d,r] for d in D for r in R)
        )
        
        # Basic Constraints
        for d in D:
            model += lpSum(x[d,r] for r in R) == 1  # Cada disciplina deve ser alocada a uma sala
        
        for r in R:
            for h in H:
                model += lpSum(x[d,r] * self.courses_data[d]['time_slots'][h]
                             for d in D) <= 1  # Evitar conflitos de horário
        
        # Room type constraints with flexibility
        for d in D:
            if self.courses_data[d]['req'] == 1:  # Precisa de laboratório
                # Se não usar laboratório, marca como wrong_lab_usage
                model += w_lab[d] == 1 - lpSum(x[d,r] for r in L)
            else:  # Precisa de sala regular
                # Se usar laboratório, marca como wrong_sala_usage
                model += w_sala[d] == lpSum(x[d,r] for r in L)
        
        # Distance constraints
        for d in D:
            for r in R:
                pref_floor = self.courses_data[d]['pref_floor']
                room_floor = self.rooms_data[r]['floor']
                model += y[d] >= x[d,r] * abs(pref_floor - room_floor)
        
        # Capacity constraints
        for d in D:
            for r in R:
                model += z[d,r] >= x[d,r] * (self.courses_data[d]['class_size'] - 
                                           self.rooms_data[r]['capacity'])
                
                # Prevent assignments that exceed capacity by more than 20%
                model += self.courses_data[d]['class_size'] <= \
                        (self.rooms_data[r]['capacity'] * 1.2 + 
                         (1 - x[d,r]) * 9999)
        
        return model, x
    
    def format_results(self, model: LpProblem, x: Dict) -> pd.DataFrame:
        """Formats optimization results into a DataFrame with room type mismatch information"""
        results = []
        
        if model.status == LpStatusOptimal:
            for d in range(len(self.courses_data)):
                for r in range(len(self.rooms_data)):
                    if value(x[d,r]) == 1:
                        # Determine if there's a room type mismatch
                        needs_lab = self.courses_data[d]['req'] == 1
                        got_lab = self.rooms_data[r]['type'] == 'lab'
                        mismatch = (needs_lab and not got_lab) or (not needs_lab and got_lab)
                        
                        results.append({
                            'Curso': self.courses_data[d]['course'],
                            'Disciplina': self.courses_data[d]['name'],
                            'Sala': self.rooms_data[r]['name'],
                            'Dia': self.courses_data[d]['day'],
                            'Horário': self.courses_data[d]['time'],
                            'Andar': self.rooms_data[r]['floor'],
                            'Tipo Sala': self.rooms_data[r]['type'].upper(),
                            'Andar Preferido': self.courses_data[d]['pref_floor'],
                            'Floor Match': self.rooms_data[r]['floor_match_dict'][self.courses_data[d]['name']],
                            'Tamanho Turma': self.courses_data[d]['class_size'],
                            'Capacidade Sala': self.rooms_data[r]['capacity'],
                            'Ocupação (%)': round(self.courses_data[d]['class_size'] / 
                                                self.rooms_data[r]['capacity'] * 100, 1),
                            'Requer Lab': needs_lab,
                            'Mismatch': mismatch
                        })
        
        return pd.DataFrame(results)
```

# 4. Funções de Análise e Exportação

```python
# Definições globais
DIAS = ['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta', 'Sábado']
HORARIOS = [
    '17h20-19h00',
    '19h00-20h40',
    '20h50-22h30'
]

def analyze_free_slots(results_df: pd.DataFrame, rooms_data: List[Dict]) -> pd.DataFrame:
    """
    Analyze and return free time slots for each room, accounting for multi-period classes
    
    Args:
        results_df (pd.DataFrame): DataFrame with current assignments
        rooms_data (List[Dict]): List of room data
    
    Returns:
        pd.DataFrame: DataFrame with actually free slots for each room
    """
    # Create all possible combinations of rooms, days and times
    todas_combinacoes = []
    for room in rooms_data:
        for dia in DIAS:
            for horario in HORARIOS:
                todas_combinacoes.append({
                    'Sala': room['name'],
                    'Tipo': room['type'].upper(),
                    'Andar': room['floor'],
                    'Dia': dia,
                    'Horário': horario
                })
    
    # Create DataFrame with all possible combinations
    todas_combinacoes_df = pd.DataFrame(todas_combinacoes)
    
    # Function to get all affected time slots for a given class
    def get_affected_slots(horario: str) -> List[str]:
        if horario == '19h00-22h30':  # 4-period class
            return ['19h00-20h40', '20h50-22h30']
        elif horario == '19h00-20h40':  # 2-period class at start
            return ['19h00-20h40']
        elif horario == '20h50-22h30':  # 2-period class at end
            return ['20h50-22h30']
        else:
            return [horario]
    
    # Expand current assignments to include all affected time slots
    expanded_occupancy = []
    for _, row in results_df.iterrows():
        affected_slots = get_affected_slots(row['Horário'])
        for slot in affected_slots:
            expanded_occupancy.append({
                'Sala': row['Sala'],
                'Dia': row['Dia'],
                'Horário': slot
            })
    
    # Convert to DataFrame
    ocupadas = pd.DataFrame(expanded_occupancy).drop_duplicates()
    
    # Merge to find free slots
    livres_df = todas_combinacoes_df.merge(
        ocupadas,
        on=['Sala', 'Dia', 'Horário'],
        how='left',
        indicator=True
    )
    
    # Keep only free slots
    livres_df = livres_df[livres_df['_merge'] == 'left_only'].drop('_merge', axis=1)
    
    return livres_df

def format_schedule_for_excel(results_df):
    """
    Formata o DataFrame de resultados de alocação em uma estrutura tabular para exportação em Excel,
    tratando corretamente disciplinas que ocupam múltiplos períodos.
    
    Args:
        results_df (pd.DataFrame): DataFrame com as alocações de salas
    
    Returns:
        pd.DataFrame: DataFrame formatado com horários combinados
    """
    # Cria uma cópia do DataFrame para não modificar o original
    df_modified = results_df.copy()
    
    # Função para ajustar o horário de exibição
    def adjust_display_time(row):
        if row['Horário'] == '19h00-22h30':
            return '19h00-22h30'  # Mantém o horário estendido como está
        return row['Horário']  # Mantém outros horários normais
    
    # Aplica o ajuste de horário
    df_modified['Horário_Display'] = df_modified.apply(adjust_display_time, axis=1)
    
    # Cria colunas para todos os horários possíveis, incluindo o período estendido
    horarios_possiveis = [
        '17h20-19h00',
        '19h00-20h40',
        '19h00-22h30',  # Adiciona o período estendido
        '20h50-22h30'
    ]
    
    # Pivota os dados usando o horário ajustado
    formatted_df = (df_modified
                   .pivot_table(index=['Sala', 'Dia'], 
                              columns='Horário_Display', 
                              values='Disciplina',
                              aggfunc=lambda x: ' | '.join(x))  # Junta disciplinas sobrepostas
                   .reindex(columns=horarios_possiveis)  # Garante a ordem correta das colunas
                   .fillna('')  # Substitui valores NaN por string vazia
                  )
    
    # Remove a coluna do período intermediário quando há aula estendida
    def clean_overlapping_slots(row):
        if row['19h00-22h30']:
            # Se há aula no período estendido, limpa os períodos intermediários
            row['19h00-20h40'] = ''
            row['20h50-22h30'] = ''
        return row
    
    # Aplica a limpeza
    formatted_df = formatted_df.apply(clean_overlapping_slots, axis=1)
    
    # Flatten the columns for easier viewing
    formatted_df.columns = [f"{col}" for col in formatted_df.columns]
    
    return formatted_df.reset_index()
```
